```python
df_val = pd.read_csv(X_val, sep='\t')
val_list = df_val['0'].to_list()
```

```python
def get_predictions(val_list, model):
    y_pred = []
    for i in val_list:
        y_i = model.predict(i.replace("\n", ""), k=-1, threshold=0.5)
        y_pred.append(y_i)
    return y_pred
```
