```python
import pandas as pd
import numpy as np
import json
```

```python
def get_data_and_save():
    articles = []
    with open("/content/arxiv-metadata-oai-snapshot.json", "r") as f:
        for l in f:
            d = json.loads(l)

            if d['categories'].split(' ')[0].startswith('cs'):
                articles.append((d['id'], d['authors'], d['title'], d['abstract'], d['categories']))

    dat = pd.DataFrame(articles)
    dat.columns = ['id', 'authors', 'title', 'abstract', 'categories']

    dat.to_csv('./data/clean_data.csv', index = False)
    return dat
```

```python
get_data_and_save()
```
